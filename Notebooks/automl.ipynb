# Library Imports

```python
import pandas as pd
from pycaret.regression import RegressionExperiment
# https://pycaret.gitbook.io/docs/get-started/quickstart#regression
```

# Read and Preprocess Data

```python
# read data
data = pd.read_csv('Begining_Dataset.csv')

# drop the na rows
data = data.dropna()

# remove columns 10 to 19 except for column 15
# 10 to 19 are the homeless count targets not features
# column 15 is the selected target for this run
drop_idx = list(range(10,15)) + list(range(16,20))
drop_cols = data.columns[drop_idx]
print(drop_cols)
data = data.drop(columns=drop_cols)

# print out some rows
data.head()
```

# Run Experiment

```python
s = RegressionExperiment()
s.setup(data,
        target="Overall_Homeless_Per_Capita",
        session_id= 222)
best = s.compare_models()
print(best)
```
