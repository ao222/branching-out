#Average Temperature Providence
Confirming average temperature comes from January Temperatures.

```python
import pandas as pd
import numpy as np

weather_data = pd.read_csv("final_weather_data.csv")
dataset = pd.read_csv("Begining_Dataset.csv")
wd_values = weather_data.values.ravel()                             # create one series from all columns
wd_values = pd.concat([pd.Series(wd_values), pd.Series([np.nan])])  # add a NaN value (because NaNs are present)
avg_temp = dataset["Average Temperature"]                           # get a series
dataset["from_table"] = avg_temp.isin(wd_values)                    # check which vals are from this table
np.sum(dataset["from_table"]) == len(dataset)                       # count the Trues
```

# Conclusion
The values in the Average Temperature column comes from the average state January temperature table.
